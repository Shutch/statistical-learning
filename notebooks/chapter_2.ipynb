In general, flexible models have high variance and low bias, and inflexible is the opposite

- The sample size n is extremely large, and the number of predictors p is small.
    - Flexible would be better as it could fit the large dataset with low chance of overfitting
- The number of predictors p is extremely large, and the number of observations n is small.
    - Inflexible to minimize the effects of variance given the low number of observations.
- The relationship between the predictors and response is highly non-linear.
    - Flexible would be better to capture the non-linear relationship (more parameters).
- The variance of the error terms, i.e. σ2 = Var("), is extremely high.
    - Inflexible would be better because a flexible model would likely overfit on data with high variance.
